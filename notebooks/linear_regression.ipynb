## Linear Regression Using Gradient Descent 

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8-whitegrid")
```

```python
np.random.seed(42)
X_feature = np.random.rand(100,1)
X_feature
```

```python
np.random.seed(43)
noise = np.random.rand(100,1)
y = 4 + 3*X_feature + noise
y.shape
```

```python
ones = np.ones((100,1))
X = np.append(arr=ones,values=X_feature,axis=1)
X
```

```python
np.random.seed(42)
theta = np.random.randn(2,1)
print(theta)
print(theta.shape)
```

```python
theta[0]
```

```python
alpha = 0.1
iterations = 1000
m = len(y)

for i in range(iterations):
    #calculate the predictions
    predictions = np.dot(X,theta)
    #calculate the error 
    error = predictions-y
    #calculate the gradient (on average the how much should we move each parameter to minimize the loss function)
    gradient = (1/m) * np.dot(X.T,error)
    #change the parameters
    theta = theta -alpha*gradient
print(theta)

    
```

### see how well our ÿ (estimation) is with the real y

```python
yhat = theta[0]+theta[1]*X_feature
fig,ax = plt.subplots(figsize=(12,8))
plt.scatter(X_feature,y,color="blue",linewidth=2,label="Actual Data")
plt.plot(X_feature,yhat,color="red",linewidth=1,label="Model Prediction (yhat)")
plt.legend()
plt.show()
plt.tight_layout()
```
